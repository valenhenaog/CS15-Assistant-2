```python
url = 'https://piazza.com/class/m5v7vm0vz5i4y6/post/m65p08up7ut4jb'

page = requests.get(url)

soup = BeautifulSoup(page.content, 'html')

print(soup.prettify())
```

```python
react_feed_divs = soup.find_all('div', class_='react_feed')
if react_feed_divs:
    react_feed_text = react_feed_divs[0].get_text()
else:
    react_feed_text = None

print(react_feed_text)
```
